```
columns_to_drop=[
    'Restaurant ID',
    'Restaurant Name',
    'Locality Verbose',
    'Switch to order menu',
    'Address'
]
df.drop(columns=columns_to_drop,inplace=True,errors='ignore')

```

ENCODING CATEGORICAL VALUES
# BINARY CATEGORICAL
## Label Encoding
* table booking
* online booking
* delivering now


```
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
x=df['Has Table booking']
df['Has Table booking']=lab.fit_transform(x)

y=df['Has Online delivery']
df['Has Online delivery']=lab.fit_transform(y)

s=df['Is delivering now']
df['Is delivering now']=lab.fit_transform(s)
```

# ORDINAL CATEGORICAL
## Ordinal Encoding
* rating text
* rating color

```
from sklearn.preprocessing import OrdinalEncoder
mab=OrdinalEncoder()
df['Rating text']=mab.fit_transform(df[['Rating text']])

df['Rating color']=mab.fit_transform(df[['Rating color']])
```

# NOMINAL CATEGORICAL
* country code = One hot encoding
* currency = one hot encoding
* city = frequency encoding
* locality = target encoding

```
from sklearn.preprocessing import OneHotEncoder,TargetEncoder
hp=OneHotEncoder()
ph=TargetEncoder()

df['Country Code']=hp.fit_transform(df['Country Code'])
df['Currency']=hp.fit_transform(df['Currency'])

df['locality']=ph.fit_transform(df['locality'])
```

# Multi label feature
* cuisines = multi label encoding

```
from sklearn.preprocessing import MultiLabelBinarizer
bs=MultiLabelBinarizer()
df['Cuisines']=bs.fit_transform(df['Cuisines'])
```
